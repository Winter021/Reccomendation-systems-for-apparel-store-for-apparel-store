Preprocessing

```python
# adding product code to transactions df
transactions = pd.merge(transactions, articles[['article_id', 'product_code', 'detail_desc']], left_on="article_id",right_on="article_id")
transactions.head()
```

```python
# dropping customer fn and fn freq
customers = customers[["customer_id","age"]]
```

```python
# handling nan in customers df
print("Total customers",customers.shape[0])
print("Customers with missing age",customers["age"].isna().sum())
print("%: ",customers["age"].isna().sum()/customers.shape[0]*100)
```

```python
customers.dropna(subset=["age"],inplace=True)
print("Total customers after drop: ",customers.shape[0])
print("Customers with missing age",customers["age"].isna().sum())
```

```python
# recommending last weeks popular purchases
transactions.t_dat = pd.to_datetime(transactions.t_dat)
last_week_transactions = transactions.loc[transactions.t_dat >= pd.to_datetime('2020-09-16')]
# top_12 = ' 0' + ' 0'.join(last_week_transactions.article_id.value_counts().to_pandas().index.astype('str')[:12])
popular_articles = last_week_transactions.groupby('product_code_x').size().reset_index(name='purchase_count')

# Sort articles by purchase count in descending order
popular_articles = popular_articles.sort_values(by='purchase_count', ascending=False).reset_index()

popular_articles_desc = articles[articles['product_code'].isin(popular_articles['product_code_x'])]
# Display the most popular articles
popular_articles_desc.head(15)[['article_id','prod_name','product_type_name','product_group_name','department_name','index_name','detail_desc']]
```

```python
# recommending Customers popular purchases
tmp = transactions.groupby(['customer_id','product_code_x'])['t_dat'].agg('count').reset_index()
tmp.columns = ['customer_id','product_code_x','ct']
customers_wise_popular_transactions = transactions[["customer_id", "product_code_x","detail_desc"]].merge(tmp,on=['customer_id','product_code_x'],how='left')
customers_wise_popular_transactions = customers_wise_popular_transactions.sort_values(['ct'],ascending=False)
customers_wise_popular_transactions = customers_wise_popular_transactions.drop_duplicates(['customer_id','product_code_x'])
customers_wise_popular_transactions = customers_wise_popular_transactions.sort_values(['ct'],ascending=False)
customers_wise_popular_transactions = customers_wise_popular_transactions[customers_wise_popular_transactions['customer_id']=='00000dbacae5abe5e23885899a1fa44253a17956c6d1c3d25f88aa139fdfc657']
customers_wise_popular_transactions.head(10)
```
